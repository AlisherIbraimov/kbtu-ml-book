# Impurity and information criterions
Information criteria are statistical measures used to evaluate and compare different models, including decision trees, based on their ability to fit the data while penalizing for model complexity. Information criteria are often used during the process of splitting nodes in decision trees to determine the best feature and split point. Two common information criteria used in this context are the Gini impurity and entropy.

## Impurity measure in classification trees

Let $\mathcal Y = \{1, \ldots, K\}$. Estimate probability of class $k$ at some node $V$ as

```{math}
    p_k = \frac 1n\sum\limits_{(\boldsymbol x_i, y_i)\in V} \mathbb I[y_i = k].
```

The main two impurity measures for classification are **Entropy** and **Gini**.

### Entropy
Entropy is a metric to measure the impurity in a given attribute. It specifies randomness in data. In a decision tree, the goal is to decrease the entropy of the dataset by creating more pure subsets of data. Since entropy is a measure of impurity, by decreasing the entropy, we are increasing the purity of the data.

__Formula of Entropy__<br>
 <br>

The general form of the Entropy for a probability distribution $p = \{p_1, p_2, ..., p_k\}$ is given by:

```{math}
:label: entropy
    \mathbb H[p] = -\sum_{k}^{}p_k \log_2\, p_k.
```

```{important}
$0log(0)=0$ for the special case when the probability $p_k$ is zero.
```

Let us imagine we have a set of cats and dogs. Let's denote $p_1$ as the fraction of examples that are cats. To measure the impurity of the sets we use Entropy.Consider $p_1$ as the fraction of examples representing cats. For instance, if $2/3$ of a sample consists of cats, then the remaining $1/3$ must represent non-cats. Let's denote $p_0$ as the fraction of examples that are not cats:


$$p_0=1−p_1.$$

The Entropy function then defined as: <br>


$$\mathbb H[p_1] = -p_1log_2(p_1) - p_0log_2(p_0) = -p_1log_2(p_1) -(1-p_1)log_2(1-p_1).$$

__Entropy of a Joint Distribution__ <br>
 <br>
 Specific conditional entropy $\mathbb H(X|Y=v)$ of $X$ given $Y=v$ :

$$\mathbb H(X|Y=v) = \mathbb -\sum_{i}^{}P(X=i|Y=v) \log_2\, P(X=i|Y=v).$$

Conditional entropy $ \mathbb H(X|Y)$ of $X$ given $Y$:

$$H(X|Y) = \sum_{i}^{}P(Y=v) \log_2\, P(X|Y=v).$$

### Gini Impurity
**Gini Impurity** measures the frequency at which any element of the dataset will be mislabelled when it is randomly labeled.

![picture](https://drive.google.com/uc?id=16ffcsDm3O4ZIRMxljhDC4WSOsvEghjGm)


 For a credit card default prediction, imagine sorting people based on age, income, credit rating, and student status. To find the best feature for the first split of the tree – the root node, evaluate how well each feature separates those who default ("yes") from those who don't ("no"). This calculation would measure the **impurity** of the split, and the feature with the lowest impurity would determine the best feature for splitting the current node. This process would continue for each subsequent node using the remaining features.

In the image above, age has minumum gini impurity, so age is selected as the root in the decision tree.

#### Mathematical definition
The Gini impurity is calculated using the following formula:

```{Note}
Note that $-\ln(1-t) \sim t$ as $t \to 0$. After replacing $-\log p_k$ by $1-p_k$ in {eq}`entropy`, the entropy turns into **Gini** impurity measure
```

$$
    \mathbb G[p] = \sum\limits_k p_k(1-p_k).
$$

where $p_k$ is the probability of class $k$.

**Q**. What is the Gini measure of the uniform distribution $p_k = \frac 1K$, $k=1,\ldots, K$?

The minimum value of the Gini is 0. This happens when the node is **pure**, this means that all the contained elements in the node are of one unique class. Therefore, this node will not be split again. Thus, the optimum split is chosen by the features with less Gini. Moreover, it gets the maximum value when the probability of the two classes are the same.


$$ \mathbb G_{\text{min}}=1–(1^2)=0 $$
$$ \mathbb G_{\text{max}}=1–(0.5^2+0.5^2)=0 $$

**Comparing Entropy and Gini**

If $K=2$, then probabilites are equal to $p$ and $1-p$ for some $p \in [0, 1]$. For this particular case the comparison of entropy and Gini measures is shown on the next picture.

````python
import plotly.graph_objects as go
import numpy as np

# Функция для расчета энтропии
def entropy(p):
    return - p * np.log2(p) - (1 - p) * np.log2(1 - p)

# Функция для расчета индекса Gini
def gini(p):
    return 2 * p * (1 - p)

# Вероятности от 0 до 1
probabilities = np.linspace(0.01, 0.99, 100)

# Рассчитываем значения энтропии и индекса Gini для каждой вероятности
entropies = [entropy(p) for p in probabilities]
gini_values = [gini(p) for p in probabilities]

# Создаем график
fig = go.Figure()

# Добавляем кривые для энтропии и индекса Gini
fig.add_trace(go.Scatter(x=probabilities, y=entropies, mode='lines', name='Entropy'))
fig.add_trace(go.Scatter(x=probabilities, y=gini_values, mode='lines', name='Gini'))

# Настройка макета с ограничением ширины
fig.update_layout(
    # title='Entropy and Gini Index Curves',
    xaxis_title='Probability',
    yaxis_title='Value',
    legend=dict(x=0, y=1, traceorder='normal', orientation='h'),
    width=600  # Задайте желаемую ширину
)

# Отображение графика
fig.show()
````

### Information Gain
**Information gain** or **IG** is a statistical property that measures how well a given attribute separates the training examples according to their target classification. Information gain is a decrease in entropy. It computes the difference between entropy before split and average entropy after split of the dataset based on given attribute values.

$$ \mathbb H (p^{\text{root}}_1) - (w^{\text{left}} \mathbb H (p^{\text{left}}_1)  + (w^{\text{right}} \mathbb H (p^{\text{right}}_1)). $$

Mututal information (aka Information Gain) of X and Y :

               I(X|Y) = H(X) - H(X|Y) = H(Y) - H(Y|X)

If X is completely uninformative about Y : IG (Y |X ) = 0

If X is completely informative about Y: IG(Y|X) = H(Y)

### Iris dataset

Decision Tree Training on **Iris Data** with Entropy and Gini Criteria:

````python
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier, export_graphviz

iris = sns.load_dataset("iris")
y = iris['species']
X = iris.drop("species", axis=1)

clf_ent = DecisionTreeClassifier(criterion='entropy', max_depth=2)
clf_ent = clf_ent.fit(X, y)
print("Entropy accuracy:", clf_ent.score(X, y))

clf_gini = DecisionTreeClassifier(criterion='gini', max_depth=2)
clf_gini = clf_gini.fit(X, y)
print("Gini accuracy:", clf_gini.score(X, y))
````

Decision Tree Visualization

````python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
X, y = iris.data, iris.target

# Function to build and plot a decision tree
def build_and_plot_tree(criterion, title):
    clf = DecisionTreeClassifier(criterion=criterion)
    clf.fit(X, y)

    plt.figure(figsize=(20, 10))
    plot_tree(clf, filled=True, feature_names=iris.feature_names,
              class_names=iris.target_names, rounded=True, proportion=True)
    plt.title(f"Decision Tree using {title}")
    plt.show()

# Building and plotting trees using Gini and Entropy

````

Gini decision tree

````python
build_and_plot_tree('gini', 'Gini Impurity')
````

Entropy decision tree

````python
build_and_plot_tree('entropy', 'Entropy')
````
