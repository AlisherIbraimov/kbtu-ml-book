# Gaussian Model
When it comes to linear regression, the Gaussian model makes the assumption that the error (the difference between the predicted and the real value)  follows a normal distribution. In other words, significant errors will be uncommon and the majority will be small, centered around zero.
<br>

$$
\begin{aligned}
y = xw + ε
\end{aligned}
$$

<br>

$ε$ is assumed to be normally distributed with zero mean and some variance $σ^{2}$. Thus, the probability of observing a specific value of $y$ for a given $x$ is described by the formula:

$$
\begin{aligned}
P(y|x, w) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{ -\frac{(y - (wx+b))^2}{2\sigma^2} } 
\end{aligned}
$$

```{admonition} Click here!
:class: tip, dropdown

$e$ — natural logarithm

$
{ -\frac{(y - (wx+b))^2}{2\sigma^2} } 
$ — to measure the probability of how much the result may deviate from the true value.
```

The likelihood function:

$$
    p( y \vert X, w) = \prod_{i=1}^n \mathcal N( x_i^{\mathsf T} w, \sigma^2) = \frac 1{(\sqrt{2\pi}\sigma)^n}\exp\Big(-\sum\limits_{i=1}^n \frac{(x_i^{\mathsf T} w - y_i)^2}{2\sigma^2}\Big).
$$

<br>
The likelihood of the observed data for a given set of model $w$ parameters is expressed as the likelihood function, $p(y∣X,w)$. If the errors have a normal distribution with variance $σ^2$, it is the product of the probability of the individual observations.

Negative log-likelihood:

$$
    \mathrm{NLL}( w) = -\log p( y \vert  X,  w)=
    \frac 1{2\sigma^2} \sum\limits_{i=1}^n ( x_i^{\mathsf T} w - y_i)^2 + \frac n2\log(2\pi\sigma)^2.
$$

<br>
A likelihood function transformation called $NLL(w)$ is used to minimize and streamline the optimization process. This function consists of an additional component that depends on the variance and the number of observations, as well as the sum of the squares of the differences between the observed and predicted values divided by the error variance.

# Laplaciaan Model
The Laplace distribution, which has more "sharp" peaks than the normal distribution, is the assumption made by the Laplacian model for the distribution of errors. This indicates that in comparison to the normal distribution, the Laplace model has a higher likelihood of both very small and very large errors.

<br>

$$
\begin{aligned}
ε = y-(wx+b)
\end{aligned}
$$

<br>

$ϵ$ obeys the Laplace distribution with location parameters ($μ$, usually 0) and scale (beta, $β$). The probability density for Laplace is expressed as:

$$
\begin{aligned}
P(y|x) = \frac{1}{2\beta} e^{ -\frac{|y-(wx+b)|}{\beta} } 
\end{aligned}
$$


```{admonition} Click here!
:class: tip, dropdown


$P(y|x)$ — the probability of getting value of dependent variable $y$, indepent variable $x$;

$β$ — The scale parameter, which defines the degree of width;

$e$ — natural logarithm

$
{ -\frac{(y - (wx+b))}{β} } 
$ —  reflects an exponential decrease in probability with an increase in the absolute value of the prediction error.

```

Negative log-likelihood:

$$
    \mathrm{NLL}(w) = \frac 1b\sum\limits_{i=1}^n \vert x_i^{\mathsf T} w - y_i\vert + \mathrm{const}?
$$


```{admonition} Click here!
:class: tip, dropdown

Negative logarithmic likelihood (NLL) uses the absolute value of the difference multiplied by $\frac 1b$, where $b$ is the scale parameter of the Laplace distribution, to determine the degree of discrepancy between the observed data and the predicted values.

```

Likelihood function will be: 

$$
    p(y \vert X, w) \propto \exp\Big(-\frac 1b\sum\limits_{i=1}^n \vert x_i^{\mathsf T} w - y_i\vert\Big) = \prod\limits_{i=1}^n \exp\Big(-\frac{\vert x_i^{\mathsf T} w - y_i\vert} b\Big).
$$


```{admonition} Click here!
:class: tip, dropdown

If the error is distributed over Laplace with a scale parameter $b$, the likelihood function $p(y∣X,w)$ describes the probability of observing data y based on a model with parameters $w$ and $X$.

```



# Bayesian Linear Regression
Bayesian Linear Regression extends the concept of linear regression by introducing the concepts of Bayesian statistics. In this model, instead of looking for fixed values for parameters, probability distributions are used.
<br><br>
## Gaussian Prior
Gaussian prior assumed that the distribution of the weights (w) follows a common pattern where they cluster around an average number (μ) and vary within a certain range (σ²). This can be represented as:

<br>

$$
\begin{aligned}
w \sim \mathcal{N}(μ, σ^{2})
\end{aligned}$$

<br>
The formula of the Bayesian approach to linear regression, where the Gaussian prior is used: 


$$
p(w|X, y) \propto p(y|X, w)p(w)
$$

<br>
The formula $p(w∣X,y)$ describes the a posteriori distribution of the parameters of the model $w$, taking into account the data $X$ and $y$. The a posteriori distribution is proportional to the product of the likelihood $p(y∣X,w)$ by the a priori distribution p(w).

To find ${\widehat w}_{\mathrm{MAP}}$, we need to maximize posterior $p(w \vert X, y)$. This is the same as to minimize

$$
    -\log p(w \vert X, y) = -\log p(y \vert X, w) - \log p(w) + \mathrm{const}.
$$

<br>
In Bayesian linear regression with a Gaussian a priori distribution, this formula represents the negative logarithm of a posteriori probability. It demonstrates that minimizing the sum of the negative logarithms of the data likelihood and the a priori parameter distribution is required to determine the a posteriori maximum (MAP) estimate of the model parameters. Constants that are independent of parameters and have no bearing on optimization are reflected in the 'const' term.

Recall that $p(y \vert X, w) = \mathcal N({Xw}, \sigma^2 I)$. According to calculations from [ML Handbook](https://academy.yandex.ru/handbook/ml/article/beta-bajesovskij-podhod-k-ocenivaniyu#primer-linejnaya-regressiya-s-l-2-regulyarizacziej-kak-model-s-gaussovskim-apriornym-raspredeleniem-na-vesa)

$$
    -\log p(y \vert X, w) - \log p(w)
    = \frac 1{2\sigma^2} \sum\limits_{i=1}^n (y_i -x_i^\intercal w)^2 + \frac 1{2\tau^2} \sum\limits_{j=1}^d w_j^2 + \mathrm{const}.
$$

```{admonition} Click here!
:class: tip, dropdown

This formula represents the negative logarithm of the joint probability. The negative logarithm of the likelihood of the data, which is the sum of the squares of the differences between the actual and predicted values of $x_i^\intercal w$ is divided by the noise variance $σ^2$. The sum of the squares of the weights $w_j$ divided by the variance of the a priori distribution $τ^2$ provides the second term, which is the negative logarithm of the a priori probability of the weights $w$. Terms that are unrelated to w and have no bearing on optimization are included in the const constant.
```

Hence,

$$
{\widehat w}_{\mathrm{MAP}} = \arg\min\limits_{w}
\bigg(\sum\limits_{i=1}^n (y_i - x_i^\intercal w)^2 + \frac{\sigma^2}{\tau^2} \Vert w \Vert_2^2\bigg).
$$

```{admonition} Click here!
:class: tip, dropdown

The sum of squares of prediction errors and weights regularized by the $\frac{\sigma^2}{\tau^2}$parameter are combined to solve the minimization problem using ${\widehat w}_{\mathrm{MAP}$. The ridge regression weights are estimated in accordance with this solution.
```

Contnuing calculations, we can find posterior:

$$
    p(w \vert X, y) = \mathcal N\Big({\widehat w}_{\mathrm{MAP}},\Big(\frac 1{\sigma^2}X^\intercal X + \frac{1}{\tau^2} I\Big)^{-1} \Big).
$$.

<br>
Following the observation of data $X$ and $Y$, the parameters of $w$ have a normal distribution around the covariance matrix, the inverse sum of the matrix ${X^T}X$ scaled by the inverse variance of noise $σ^2$, the unit matrix scaled by the inverse variance of the a priori distribution $τ^2$, and the estimate of the maximum a posteriori value of ${\widehat w}_{\mathrm{MAP}}$.

## Laplacian Prior
Laplace prior model's weights are spread around mean(μ, usually 0), and the spred is controlled by the scale parameter, beta(β) This could be represented as: 

<br>

$$
\begin{aligned}
P(w) = \frac{1}{2β} e^{-\frac{|w-μ|}{b} }
\end{aligned}
$$

```{admonition} Click here!
:class: tip, dropdown


$P(w)$ — Probability of $w$;

$
\frac{1}{2β}
$ — ensures that the total area under the distribution curve is 1, which is a prerequisite for probability functions.

```

Consinder the following prior: $w_1, \ldots, w_d$ are i.i.d laplacian random variables with parameter $\lambda$. Then

$$
    p(w) = \prod\limits_{j=1}^d \mathrm{Laplace}(\lambda) = 
    \prod\limits_{j=1}^d \frac\lambda 2e^{-\lambda|w_j|} = \frac {\lambda^d}{2^d}
    \exp\Big(-\sum\limits_{j=1}^d\lambda|w_j|\Big).
$$

<br>
The probability of the vector of weights $w$, which is the product of the probabilities for each weight $w_j$ distributed by Laplace, is displayed by the formula for $p(w)$.

The likelihood $p(y \vert X, w)$ is still 
$\mathcal N({Xw}, \sigma^2 I)$. Then

$$
-\log p( w \vert X, y) = \frac 1{2\sigma^2} \sum\limits_{i=1}^n (y_i - x_i^{\mathsf{T}} w)^2 +  \lambda\sum\limits_{j=1}^d \vert w_j\vert + \mathrm{const}.
$$

<br>
The likelihood of the data (as in the Gaussian model) and the a priori distribution of weights $w$ (from the Laplace distribution) are combined to get the negative logarithm of the joint probability. The regularization of the Lasso is achieved by summing the difference of squares between the observed and predicted values and multiplying the weights' absolute values by the parameter $λ$.

Hence, maximum a posteriori estimation is

$$
{\widehat w}_{\mathrm{MAP}} = \arg\min\limits_{w}
\Big(\Vert {Xw} - y\Vert_2^2 + 2\sigma^2 \lambda \Vert w \Vert_1\Big).
$$


<br>
The task of minimizing the total squared errors and weight absolute values, regularized by a coefficient proportional to $λ$, is the maximum a posteriori estimation of ${\widehat w}_{\mathrm{MAP}}$. This is the same as using linear regression to solve the Lasso problem.

# Example with Syntethic Data

````python
%%capture
import math
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
!pip install mpld3
import mpld3
mpld3.enable_notebook()

x = np.linspace(0.0,10.0, num=100)
a = 4.0
b = -3.5
c = 0.0
y = (a*(x**2)) + (b*x) + c

#   let's add noise to the data
#   np.random.normal(mean, standardDeviation, num)
noise = np.random.normal(0, 10., 100)
y = y+noise
````

````python
#   Here we can see the placement of the data present in our synthetic datatset

sns.scatterplot(x=x,y=y,color="blue",alpha=0.8)
plt.legend(['Data'])
plt.show()
````

````python
#   Define a function to calculate the log likelihood required for the Guassian aproach

def calcLogLikelihood(guess, true, n):
    error = true-guess
    sigma = np.std(error)
    f = ((1.0/(2.0*math.pi*sigma*sigma))**(n/2))* \
        np.exp(-1*((np.dot(error.T,error))/(2*sigma*sigma)))
    return np.log(f)
````

````python
#   Define my function which will return the objective function to be minimized

def myFunction(var):
    #   load my  data
#     [x, y] = np.load('myData.npy')
    [x_, y_] = [x,y]
    yGuess = (var[2]*(x_**2)) + (var[1]*x_) + var[0]
    f = calcLogLikelihood(yGuess, y, float(len(yGuess)))
    return (-1*f)
````

````python
%%capture
#   The line above required to omit unnecessary outputs
#   Let's pick some random starting points for the initial 'guesstimate' of the function required for optimization    

nvar = 3
var = np.zeros(nvar)
var[0] = -15.5
var[1] = 19.5
var[2] = -1.0

#   let's maximize the likelihood (minimize -1*max(likelihood)

from scipy.optimize import minimize
res = minimize(myFunction, var, method='BFGS',
                options={'disp': True})
````

>BFGS — it is an algorithm that determines the descent direction by preconditioning the gradient with curvature information. It is a good choice because of its efficiency in optimizing smooth, unconstrained objective functions like those we have in our example. BFGS algorithm belongs to the quasi-Newton methods, which are iterative optimization algorithms that aim to find the minimum of a function without explicitly computing its derivatives

````python
#   perform least squares fit using scikitlearn
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
model = Pipeline([('poly', PolynomialFeatures(degree=2)),
    ('linear', LinearRegression(fit_intercept=False))])

model = model.fit(x[:, np.newaxis], y)
coefs = model.named_steps['linear'].coef_
y_pred = model.predict(x[:, np.newaxis])
````

````python
#   let's see the metrics for train data

print("Train")
print("r-score:", model.score(x[:, np.newaxis],y))
````

````python
#   And the same thing for test data

print("Test")
print("r-score:", model.score(x[:, np.newaxis],y_pred))
print("MSE:",np.mean((y_pred - y) ** 2))
````

````python
#   Let's see how our predicted values from the model above correspond to the predictions made by simple regression

sns.regplot(x=x[:, np.newaxis],y=y,color="blue",line_kws=dict(color="red"),order=2,ci=None)
plt.plot(x,y_pred,color="lime",linestyle=(5,(10,3)))
plt.legend(['Data','Least Square','Negative Log Likelihood'])
plt.show()
````

````python
#   Define a function to calculate the log likelihood required for the Guassian aproach

def calcLogLikelihood(guess, true, n):
    error = true - guess
    sigma = np.std(error)
    f = ((1.0/(2.0*math.pi*sigma*sigma))**(n/2))*np.exp(-1*((np.dot(error.T, error))/(2*sigma*sigma)))
    return np.log(f)
````

````python
#   Define my function which will return the objective function to be minimized

def myFunction(var):
    [x, y] = [df['fixed acidity'],df['pH']]
    yGuess = (var[2]*(x**2)) + (var[1]*x) + var[0]
    f = calcLogLikelihood(yGuess, y, float(len(yGuess)))
    return (-1*f)
````

````python
%%capture   
#   The line above required to omit unnecessary outputs
#   Let's pick some random starting points for the initial 'guesstimate' of the function required for optimization    

nvar = 3
var = np.zeros(nvar)
var[0] = -0.06
var[1] = -1.8
var[2] = 15

#   let's maximize the likelihood (minimize -1*max(likelihood)

from scipy.optimize import minimize
res = minimize(myFunction, var, method='BFGS',
                options={'disp': True})
````

````python
#   perform least squares fit using scikitlearn

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
model = Pipeline([('poly', PolynomialFeatures(degree=2)),
    ('linear', LinearRegression(fit_intercept=False))])

#   Here we need to transform fixed acidity column into 1-dimensional list required for the fuction to work properly
#   Also, here we are making the predictions using our model by trying to find y_pred

model = model.fit(df['fixed acidity'].values.reshape(-1, 1), df['pH'])
coefs = model.named_steps['linear'].coef_
y_pred = model.predict(df['fixed acidity'].values.reshape(-1, 1))
````
